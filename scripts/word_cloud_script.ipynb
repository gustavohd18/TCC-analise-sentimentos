```python
#Todos os tweets 3700
tweets = pd.read_excel('patch', index_col=0,  engine='openpyxl')
```

```python
#filter by sentimental
datasetFinalNeutro = tweets[tweets['emotion'] == 'neutro']
```

```python
#text cloud
text = " ".join(str(fulltext) for fulltext in datasetFinalNeutro.text)

text = text.lower()
```

```python
#word cloud image
wordcloud = WordCloud().generate(text)
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=2000, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
