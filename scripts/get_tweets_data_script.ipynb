```python
# Here you need pass your key and secret provide by twitter developer
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
```

```python
# Here we set authentification with apy.

auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tw.API(auth, wait_on_rate_limit=True,wait_on_rate_limit_notify=True)

public_tweets = api.home_timeline()
```

```python
# Here we can make a request provide a lot hashtag and combine with logic like or and
query = '(#vacina) -filter:retweets'
cursor_tweets = tw.Cursor(api.search,
            q=query,lang = 'pt',tweet_mode='extended').items(1200)
```

```python
cursor_tweets
```

```python
#Here que create a dictionary to datas
tweets_dict = {}
tweets_dict = tweets_dict.fromkeys(['created_at', 'full_text','retweet_count', 'id', 'favorite_count', 'user_name'])
```

```python
# storage data inside object
for tweet in cursor_tweets:
    for key in tweets_dict.keys():
        try:
            twvalue = tweet._json[key]
            tweets_dict[key].append(twvalue)
        except KeyError:
            twvalue = ""
            if(tweets_dict[key] is None):
                tweets_dict[key] = [twvalue]
            else:
                tweets_dict[key].append(twvalue)
        except:
            tweets_dict[key] = [twvalue]
        #print("tweets_dict[key]: {} - tweet[key]: {}".format(tweets_dict[key],  twvalue))
```

```python
dfTweets = pd.DataFrame.from_dict(tweets_dict)
```

```python
dfTweets.head()
```

```python
df2_tidy = dfTweets.rename(columns = {'full_text': 'text', 'retweet_count': 'retweets', 'favorite_count': 'favorites', 'user_name': 'username'}, inplace = False)
```
